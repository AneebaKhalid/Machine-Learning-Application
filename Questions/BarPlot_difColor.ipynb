```python
#Packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('Participant 0-RoadSet1-16.csv', sep=";")
```

```python
df
```

```python

ax = df.plot(kind='bar',x='Time1',y='Milliseconds',figsize=(40,40))
ax.axhline(2000, color="red", linestyle="--")
ax.axhline(1000, color="green", linestyle="-")
labels=df['Name1'].values+"-"+df['Name2'].values
ax.set_xticklabels(labels)
```

```python
vr = df[(df['Name1'] == ' FRD')]
```

```python
df.Name1.unique()
```

```python
#FRD = 0, Road =1 , VRCar =2, GSDRailing7 = 3, GSDRailingBase1 = 4, Rail4R = 5
df.loc[df['Name1'] == ' FRD', 'Name1_int'] = 0
df.loc[df['Name1'] == ' Road', 'Name1_int'] = 1
df.loc[df['Name1'] == ' VRCar', 'Name1_int'] = 2
df.loc[df['Name1'] == ' GSDRailing7', 'Name1_int'] = 3
df.loc[df['Name1'] == ' GSDRailingBase1', 'Name1_int'] = 4
df.loc[df['Name1'] == ' Rail4R', 'Name1_int'] = 5
```

```python
vrcar = df[(df.Name1_int)==2]
```

```python
vr_2000 = vrcar[vrcar.Milliseconds >= 2000]
```

```python
vr_2000
```

```python
pl = vrcar.plot(kind='bar',x='Time1',y='Milliseconds',figsize=(40,40))
pl.axhline(2000, color="red", linestyle="--")
pl.axhline(1000, color="green", linestyle="-")
labels=df['Name1'].values+"-"+df['Name2'].values
pl.set_xticklabels(labels)
```

```python
#Bar Plots
plt.figure(figsize=(15,8))
plt.bar(df.Time1,df.Milliseconds, color='g')
plt.axhline(2000, color="red", linestyle="--")
plt.axhline(1000, color="green", linestyle="-")
labels=df['Name1'].values+"-"+df['Name2'].values
```

```python
plt.figure(figsize=(40,40))
clrs = ['red' if (x > 2000) else 'grey' for x in df.Milliseconds ]
plt.bar(df.Time1, df.Milliseconds, color=clrs, width=0.4)
plt.show()
```
