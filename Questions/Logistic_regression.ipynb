```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
from sklearn.model_selection import cross_val_score, KFold
from sklearn.model_selection import GridSearchCV
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

df = pd.read_csv('dataset.csv')
deu = df[df['iso'] == "FRA"]

deu_filled = deu.fillna(0)
#deu_filled.columns
```

```python

#Import Library
from sklearn.linear_model import LogisticRegression
#
#Assumed you have, X (predictor) and Y (target) for 
X = deu_filled[['pk_norm', 'pk_dis', 'pk_fin' ,'govcris','turnover', 'election','gdppeak','election_year', 'govvote', 'oppvote', 'frac', 'partycount',
       'right', 'left', 'extr', 'protests']]
y = deu_filled.crisisJST

model = LogisticRegression()

sfs = SFS(model,
           k_features=14,
           forward=True,
           floating=False,
           verbose=2,
           scoring='accuracy',
           cv=5)

sfs = sfs.fit(X, y)
fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
plt.show()
print('Selected features:', sfs.k_feature_idx_)

#Training data set and x_test(predictor) of test_dataset
#X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,y,test_size=0.5)


kf = KFold(n_splits=10)
score = cross_val_score(knn, X,y, cv=kf, scoring='accuracy')


#fittedModel = model.fit(X_Train,Y_Train)
#predictions = fittedModel.predict(X_Test)


#Equation coefficient and Intercept
print('Coefficient: \n', model.coef_)
print('Intercept: \n', model.intercept_)
```
